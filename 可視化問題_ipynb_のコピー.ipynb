```python
df_1 = pd.read_csv('hist_data.csv')
df_1.head()
```

凡例の表示について<br>
legend(label(前もって代入しておく),loc = '位置を決める基準点(upper(lower) right(left))', bbox_to_anchor = (凡例boxの場所))

今回は loc = 'upper right' , bbox_to_anchor =(1,1)より箱の右上が(1.1)の位置に来るようにboxが表示された。

image.png

```python

#複数データでヒストグラムを作成するときは"[df_1['aデータ'] , df_1['bデータ']]"のようにカラムごとに記述する。
#label = ['データの表示名']；表示名を指定
#色の名称はその場で調べよう
# histtype='barstacked'；複数データを重ねて表示

plt.hist(df_1['aデータ'] ,color = 'gold' , histtype='barstacked' , bins = 35 , alpha = 0.8)
plt.hist(df_1['bデータ'] ,color = 'green' , histtype='barstacked' , bins = 35 , alpha = 0.3)
labels = ["aデータ" , "bデータ"]
#凡例の表示

plt.legend(labels ,loc = 'upper right' , bbox_to_anchor = (1,1)  )

plt.title('aデータとbデータの温度に関する度数分布')#タイトル
plt.xlabel('温度(℃)')#x軸ラベル
plt.ylabel('度数')#y軸ラベル
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_2.png?raw=1)

```python
df_2 = pd.read_csv('scatter_data.csv')
df_2.head()
```

```python
#やる事
#データをclassごとに分ける
#それぞれで散布図を作る
#分けたデータの散布図を一枚に重ね合わせる

#データをclassごとに分ける
mask_class_0 = df_2['class'] == 0
mask_class_1 = df_2['class'] == 1
mask_class_2 = df_2['class'] == 2
df_class_0 = df_2[mask_class_0]
df_class_1 = df_2[mask_class_1]
df_class_2 = df_2[mask_class_2]
```

```python
#それぞれで散布図を作る
#分けたデータの散布図を一枚に重ね合わせる
plt.scatter(df_class_0['length (cm)'] , df_class_0['width (cm)'] ,  color = 'r')
plt.scatter(df_class_1['length (cm)'] , df_class_1['width (cm)'] ,  color = 'b')
plt.scatter(df_class_2['length (cm)'] , df_class_2['width (cm)'] ,  color = 'g')
plt.grid()
plt.title('クラス0～2のlengthとwidthに関する散布図')#タイトル
plt.xlabel('length(cm)')#x軸ラベル
plt.ylabel('width(cm)')#y軸ラベル
labels = ["class0" , "class1" , "class2"]
plt.legend(labels ,loc = 'upper right' , bbox_to_anchor = (1,1)  )
```

使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_3_2.png?raw=1)

```python
df_3 = pd.read_csv('line_data.csv')
df_3.head()
```

```python
plt.plot(df_3['期間'] , df_3['件数'] , marker='.')#marker = '.'；グラフ上にスポットを表示できる
plt.title('期間ごとの件数の推移')#タイトル
plt.xlabel('期間')#x軸ラベル
plt.ylabel('件数')#y軸ラベル
#plt.xticks();x軸にメモリを追加(なぜかy軸にも追加された。.ysticks()でy軸の表示名やメモリはいじれる)
#plt.xticks(rotation = 45);x軸ラベルを45度回転
plt.xticks(rotation = 45)
plt.grid()#メモリ線に沿ってマスを作る(gridを表示する)
plt.show()

#棒グラフ
#plt.bar(x, height, オプション)
plt.bar(df_3['期間'] , df_3['費用'])
plt.grid()
plt.xticks(rotation = 45)
plt.title('期間ごとの費用の推移')#タイトル
plt.xlabel('期間')#x軸ラベル
plt.ylabel('費用(円)')#y軸ラベル
```
